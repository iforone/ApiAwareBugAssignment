```python
import numpy as np
pd.set_option("display.max_columns", None)


df = pd.read_csv ('./../data/output/del/jdt_direct.csv')


total = 0 # len(df[:6000])
count = 0
authors = {}

last_assignees= []
for index, row in df[:6000].iterrows():    

    ht = 0
    at = 0
    ct = 0
    for i in range(1, 11):
        ht += row['history_at_' + str(i) + '_v']
        ct += row['code_at_' + str(i) + '_v']
        at += row['api_at_' + str(i) + '_v']
    if ht == 0:
        ht = 1
    if at == 0:
        at = 1
    if at == 0:
        at = 1 
    
    total_scores = {}
    for i in range(1, 11):
        c = row['history_at_' + str(i)]
        v = row['history_at_' + str(i) + '_v']
        
        # damp the effect of last assignees
        p = 1
        p_reverse = 1
        for value in last_assignees[-3:]:
            if value == c and c != 'Martin Aeschlimann':
                p = 1 / (p_reverse * 10)
            p_reverse += 1
        
        total_scores.update({c: total_scores.get(c, 0) + (v / ht)})
        c = row['code_at_' + str(i)]
        v = row['code_at_' + str(i) + '_v']
        
        # damp the effect of last assignees
        p = 1
        p_reverse = 1
        for value in last_assignees[-3:]:
            if value == c and c != 'Martin Aeschlimann':
                p = 1 / (p_reverse * 10)
            p_reverse += 1
        
        total_scores.update({c: total_scores.get(c, 0) + (v / ct) })
        c = row['api_at_' + str(i)]
        v = row['api_at_' + str(i) + '_v']
        
        
        # damp the effect of last assignees
        p = 1
        p_reverse = 1
        for value in last_assignees[-3:]:
            if value == c and c != 'Martin Aeschlimann':
                p = 1 / (p_reverse)
            p_reverse += 1
        
        total_scores.update({c: total_scores.get(c, 0) +  (v / at) / 3 })

    sort = sorted(total_scores.items(), key=lambda x: x[1], reverse=True)
    
    # divide to each developer
    last_assignees.append(row['assignees'])
    
    if row['assignees'] not in authors:
        authors[row['assignees']] = {'total': 0, 'count_differently': 0,'count': 0, 'count_history': 0, 'count_code': 0, 'count_api': 0}
    
    authors[row['assignees']]['total'] += 1
    if row['assignees'] == row['history_at_1']:
        authors[row['assignees']]['count_history'] += 1
    if row['assignees'] == row['code_at_1'] or row['assignees'] == row['code_at_2']:
        authors[row['assignees']]['count_code'] += 1
    if row['assignees'] == row['api_at_1']:
        authors[row['assignees']]['count_api'] += 1
    if 1 == row['at_1']:
        authors[row['assignees']]['count'] += 1
    if sort[0][0] == row['assignees']:
        authors[row['assignees']]['count_differently'] += 1

count = 0
total = 0
for i, author in authors.items():
    author['score_history'] = author['count_history'] * 100 / author['total']
    author['score_code'] = author['count_code'] * 100 / author['total']
    author['score_api'] = author['count_api'] * 100 / author['total']
    author['score'] = author['count'] * 100 / author['total']
    author['score_differently'] = author['count_differently'] * 100 / author['total']
    count += author['count_code']
    total += author['total']

    
print(count * 100 / total)
x = pd.DataFrame.from_dict(authors)
x
```
