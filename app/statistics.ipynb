```python
df = pd.read_csv ('./../data/output/jdt_direct31-Jan-2022_06-06-31.csv')


total = len(df[:1049])
count = 0
for index, row in df[:1049].iterrows():
    best_history_candidates = [row['history_at_1'], row['history_at_2'], row['history_at_3'], row['history_at_4']]
    best_code_candidates = [row['code_at_1'], row['code_at_2'], row['code_at_3'], row['code_at_4']]
    best_api_candidates = [row['api_at_1'], row['api_at_2'], row['api_at_3'], row['api_at_4']]
    
    best_total_scores = {row['history_at_1']: 10, row['history_at_2']: 9, row['history_at_3']: 8, row['history_at_4']: 7}

    for bassbest_history_candidate in best_history_candidates:
        for i in range(1, 11):
            if row['code_at_' + str(i)] == bassbest_history_candidate:
                best_total_scores.update({bassbest_history_candidate: best_total_scores.get(bassbest_history_candidate, 0) + (1/3 * (11 - i))})
            if row['api_at_' + str(i)] == bassbest_history_candidate:
                best_total_scores.update({bassbest_history_candidate: best_total_scores.get(bassbest_history_candidate, 0) + (0 * (11 - i))})
            pass
    
    myList = sorted(best_total_scores.items(), key=lambda x: x[1], reverse=True)
    if len(myList) == 0:
        if row['history_at_1'] == row['assignees']:
            count += 1
    elif myList[0][0] == row['assignees']:
        count += 1
    
    
    #print(row['assignees'])
    #print(best_history_candidates)
    #print(best_total_scores)
    #print(myList)
    #print('OKAY')
    
print(count)
print(total)
print(count * 100 / total)
```
